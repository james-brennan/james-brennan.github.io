## The laplace approximation


The laplace approximation therefore turns the integration task primarily into an optimisation task -- once you have the mode and curvature of the posterior, the integration is quick.



```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
def normGammapdf(x1, x2, nmean, nstd, a, b):
    """
    return pdf of normal x gamma distribution
    """
    normal = scipy.stats.beta(nmean, nstd).pdf(x1)
    gamma = scipy.stats.beta(a, b).pdf(x2)
    return normal * gamma
```

```python
X1, X2 = np.meshgrid(np.linspace(0, 1, 100), np.linspace(0, 1, 100))
```

```python
pdf = normGammapdf(X1, X2, 4,3, 5,9)
```

```python
plt.contour(X1, X2, pdf, cmap='Blues')
```

Of course generally we're not going to have nice analytic expressions for real problems where a Laplace approximation might be needed.
